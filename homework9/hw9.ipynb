# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_and_get_values():
    
    x = np.linspace(0, 10, 10) 
# domain
    f_x = x**2  # f(x) = x^2
    g_x = np.sqrt(x)  # g(x) = sqrt(x)
# functions
    
    plt.figure(figsize=(10, 8))
    plt.plot(x, f_x, label='f(x) = x^2', color='blue', linestyle='--')
    plt.plot(x, g_x, label='g(x) = x', color='red', linestyle='-')

    plt.xlabel('x values', fontsize=12)
    plt.ylabel('y values', fontsize=12)
    plt.title('plots of f(x) = x^2 and g(x) = x', fontsize=20)
    plt.legend()

    plt.show()
# plot
    

    return f_x.tolist(), g_x.tolist()
# return f(x) and g(x) as lists

f_values, g_values = plot_and_get_values()

print(f_values)
print(g_values)
# get lists
```

```python
# tried to make this a module but it didnt work oops
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_function():

    x = np.linspace(-10, 10, 100)
    y = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x, y)  
# domains for x and y

    Z = X**2 + Y**2
# function

    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
# figure and 3d axis

    surface = ax.plot_surface(X, Y, Z, cmap='gnuplot2', edgecolor='none')
#plot
    
    fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)
# "legend" color bar

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.set_title('z(x, y) = x^2 + y^2')
#labels

    plt.show()

plot_3d_function()
```
